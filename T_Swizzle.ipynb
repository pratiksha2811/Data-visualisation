```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
ts_filepath = "/content/spotify_taylorswift.csv"

# Read the file into a variable fifa_data
ts_data = pd.read_csv(ts_filepath)
```

```python
ts_data.head()
```

```python
ts_data.tail()
```

```python
ts_data.info()
```

```python
plt.figure(figsize=(24,6))
sns.barplot(x=ts_data['album'],y=ts_data['popularity'])
plt.title("Album's popularity:",fontsize=13)
plt.show()
sns.set_style("dark")
```

```python
corr = ts_data.corr()
plt.figure(figsize=(24, 6))
sns.heatmap(corr, annot=True)
```

```python
ts_data['album']=pd.to_numeric(ts_data['album'],errors='coerce')
```

```python
plt.figure(figsize=(24, 6))
sns.lineplot(data=ts_data['popularity'],label="Popularity")
plt.title("Popularity")
```

```python
sns.displot(x=ts_data['popularity'], kde=True, color='#a70ad5')
plt.title('Popularity Distribution')
```

```python
sns.set_theme(color_codes=True)
plt.figure(figsize=(18, 6))
sns.regplot(x='acousticness', y='energy', data=ts_data)
plt.title('Acousticness vs Energy')
```
